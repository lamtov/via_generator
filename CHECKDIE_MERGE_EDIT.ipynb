```python
import logging
logging.basicConfig(filename='log_check_die.log',level=logging.DEBUG)
import requests
import sys
from concurrent.futures import ThreadPoolExecutor

input_file=sys.argv[1]
from pathlib import Path
my_file = Path(input_file+".txt")
if not my_file.is_file():
    print("FILE " +input_file+".txt" +" DOES NOT EXIST!")
    logging.error("FILE " +input_file+".txt" +" DOES NOT EXIST!")
number_process=int(sys.argv[2])

file = open(input_file+".txt", "r", encoding='utf-8')
line_count = 0
for line in file:
    if line != "\n":
        line_count += 1
file.close()
lines_per_file = int(line_count/number_process)
smallfile = None
with open(input_file+'.txt', encoding='utf-8') as bigfile:
  i=0
  for lineno, line in enumerate(bigfile):
      if lineno % lines_per_file == 0:
          if smallfile:
              smallfile.close()
          i=i+1
          small_filename = 'small_file_{}.txt'.format(input_file + str('_')+ str(i))
          smallfile = open(small_filename, "w", encoding='utf-8')
      print(line )
      smallfile.write(line)
  if smallfile:
      smallfile.close()
headers_text = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:82.0) Gecko/20100101 Firefox/82.0",
    "Accept": "application/json, text/javascript, */*; q=0.01",
    "Accept-Language": "vi-VN,vi;q=0.8,en-US;q=0.5,en;q=0.3",
    "X-Requested-With": "XMLHttpRequest"
}

def check_die_za(file_name):
    logging.debug(file_name)
    with open(file_name, encoding='utf-8') as namefile:
        diefile = open("die" + "_" + file_name, "w", encoding='utf-8')
        for index, line in enumerate(namefile):
            line=line.strip().replace('\n', '')
            logging.debug('line%s',line.encode("UTF-8"))
            url = str("https://www.webmail.co.za/reg/addresscheck.php?email=" + line + "@webmail.co.za")
            r = requests.get(url, headers=headers_text)
            if r.text == '1':
                logging.debug('%s',str(str(index) + ": " + str(line) + " DIE").encode("UTF-8"))
                print(str(str(index) + ": " + str(line) + " DIE").encode("UTF-8"))
                diefile.write(line+'\n')
            else:
                logging.debug('%s',str(str(index) + ": " + str(line) + " IN_USED").encode("UTF-8"))
                print(str(str(index) + ": " + str(line) + " IN_USED").encode("UTF-8"))
        diefile.close()
for i in range(1, number_process + 1):
    executor = ThreadPoolExecutor(1)
    executor.submit(check_die_za, "small_file_" + input_file + str('_') + str(i) + ".txt")


```

```python
file1 = open('BTCTN.txt', encoding="utf8")
Lines1 = file1.readlines()
output = open('BTCTN_screen.txt', 'w', encoding='utf-8')
import re
heroRegex = re.compile (r'screen_name\:.*')
for Lines in [Lines1]:
    for line in Lines[:]:
        line=line.lower()
        line=heroRegex.search(line).group().split(':')[1]
        # line = line.split(":")[2]
        line= line.replace('\n', '')
        # print(line.isnumeric() )
        if (not line.isnumeric() ) and (line != "invalid"):
            line=line.replace(" ", "")

            line=line.replace("@","").replace("'", '-').replace('!', '-').replace('#', '-').replace(':', '-').replace(';', '-').replace('/', '-').replace('-','')
            # line= unicode(line, "utf-8")
            line = line
            line=line+'\n'
            output.writelines(line)
output.close()
```

```python
import os
cur_dir = os.getcwd()
database_dir = cur_dir+'/via_database/domain_database'
def sort_and_save(file_path):

    output = open(file_path[:-4]+'_sort.txt', 'w', encoding='utf-8')
    origin_file=open(file_path, encoding="utf8")
    Lines=origin_file.readlines()
    print(Lines)
    Lines.sort(key = lambda x: x.split('.')[-1])
    print(Lines)
    for line in Lines:
        output.writelines(line)

    output.close()
    
def sort_by_size(file_path):
    output = open(file_path[:-4]+'_size.txt', 'w', encoding='utf-8')
    origin_file=open(file_path, encoding="utf8")
    Lines=origin_file.readlines()
    print(Lines)
    
    end=Lines[0].split('.')[-1]
    _list=[Lines[0]]
    for i in range(1,len(Lines)):
        if Lines[i].split('.')[-1]!=end:
            _list.sort()
            _list.sort(key = len)
            for line in _list:
                output.writelines(line)
                
            end=Lines[i].split('.')[-1]
            _list=[Lines[i]]
        else:
            _list.append(Lines[i])
            
        if i==len(Lines)-1:
            _list.sort()
            _list.sort(key = len)
            for line in _list:
                output.writelines(line)

    output.close()
    
#sort_and_save(database_dir+"/small_popular_domain.txt")
#sort_by_size(database_dir+"/small_popular_domain_sort.txt")

sort_and_save(database_dir+"/all_email_domain.txt")
sort_by_size(database_dir+"/all_email_domain_sort.txt")

sort_and_save(database_dir+"/medium_popular_domain.txt")
sort_by_size(database_dir+"/medium_popular_domain_sort.txt")
```
